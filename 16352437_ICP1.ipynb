```python
missing_values_count = df.isnull().sum()
total_entries = len(df)
missing_percentage = (missing_values_count / total_entries) * 100
missing_summary = pd.DataFrame({
    'Missing Values': missing_values_count,
    'Percentage Missing': missing_percentage
})

print("\nPercentage of missing values in each column:\n", missing_summary)
```

# **Replacing missing values with the mean values of the column**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
num_col= df.select_dtypes(include='number').columns
df[num_col] = df[num_col].apply(pd.to_numeric, errors='coerce')

# mean values for numeric columns
mean_val = df[num_col].mean()

# Replaced missing values with the mean
df[num_col].fillna(mean_val, inplace=True)

# Results
print("DataFrame after replacing missing values with column means:\n", df.head())
print("\nMissing values after replacement:\n", df.isnull().sum())
```
